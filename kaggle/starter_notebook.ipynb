# 🥃 WhiskyDB — Fine Spirits Sample: Starter Notebook

Loads the free WhiskyDB sample (spirits, distilleries) and the fully open cask & flavor taxonomies.

- Dataset: [ahtiticheamine/whiskydb-fine-spirits-sample](https://www.kaggle.com/datasets/ahtiticheamine/whiskydb-fine-spirits-sample)
- Full dataset (1,290+ spirits · 3,200+ distilleries · 20,000+ auction benchmarks): [whisky-database.pages.dev](https://whisky-database.pages.dev/)
- Docs: [GitHub — WhiskyyDB/whisky-database](https://github.com/WhiskyyDB/whisky-database)

```python
import os
import pandas as pd

# locate the dataset wherever Kaggle mounted it
BASE = "/kaggle/input/whiskydb-fine-spirits-sample"
if not os.path.isfile(os.path.join(BASE, "spirits.csv")):
    for root, _dirs, files in os.walk("/kaggle/input"):
        if "spirits.csv" in files:
            BASE = root
            break
print("reading from", BASE)

spirits = pd.read_csv(f"{BASE}/spirits.csv")
distilleries = pd.read_csv(f"{BASE}/distilleries.csv")
casks = pd.read_csv(f"{BASE}/casks.csv")
flavors = pd.read_csv(f"{BASE}/flavors.csv")

print(f"{len(spirits)} spirits · {len(distilleries)} distilleries · {len(casks)} cask styles · {len(flavors)} flavor descriptors")
spirits.head(10)
```

```python
# Spirit types and ABV distribution in the sample
spirits.groupby("type").agg(n=("spirit_id", "count"), mean_abv=("abv", "mean")).sort_values("n", ascending=False)
```

```python
# Where do the sample distilleries come from?
distilleries.groupby(["country", "region"]).size().rename("distilleries").reset_index()
```

```python
# The open 3-tier cask taxonomy (CC-BY-4.0) — usable in any project with attribution
casks[["category", "sub_type", "specific_style", "wood_species"]]
```
